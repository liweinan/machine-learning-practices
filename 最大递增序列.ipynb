```python
import collections
import random


class MultiValuedDict:
    def __init__(self):
        self.dict = {}

    def put(self, lst):
        key = len(lst)
        if not self.dict.get(key):
            self.dict[key] = []
            print("初始化key的list：", self.dict[key])

        self.dict[key].append(lst)
        print("插入的新列表：", self.dict.get(key))

    def dump(self):
        print("字典里的内容：")
        self.dump_dict(self.dict)

    def dump_dict(self, dict):
        for k, vs in dict.items():
            print(k, ": ", vs)

    def find_max(self):
        sorted_dict = collections.OrderedDict(sorted(self.dict.items()))
        print("排序后的dict：")
        self.dump_dict(sorted_dict)
        # 返回最大长度的列表们
        # last one is max one because items are sorted
        last_one = len(list(sorted_dict.keys())) - 1
        return list(sorted_dict.values())[last_one]


def longest_increasing_subseq():
    # 创建一个里面有10个随机数字的list
    random_list = []
    for i in range(10):
        random_list.append(random.randrange(1, 101, 1))
    print("原始列表: ", random_list)

    # marker用来标记上次一找到的sublist的位置
    marker = -1
    found = False

    target_list = []
    # 保存找到的所有递增子序列
    saved_sublists = MultiValuedDict()

    # 算法的思路如下：
    # 查看下一个元素，如果比之前的子序列里面的元素小，那么之前的子序列回退到比这个元素小的位置，然后在子序列里加入这个元素。
    for next_item in random_list:
        print("当前的递增序列：", target_list)
        print("下一个要添加的值：", next_item)
        # 使用enumerate函数，可以在遍历list的同时，得到每个元素的index值。
        for idx, _ in enumerate(target_list):
            print("当前的递增序列里要比对的值：", _)
            if _ > next_item and found is False:
                print("要添加的值更小！")
                print("元素位置：", idx)
                # 找到了之前的子序列里第一个大于下一个item value的item位置。
                marker = idx
                found = True
        if marker >= 0:
            target_list = target_list[:marker]
            marker = -1
            found = False
        target_list.append(next_item)
        saved_sublists.put(target_list)

    print("所有的最大递增子序列：")
    for _list in saved_sublists.find_max():
        print(_list)


# 运行十次
for i in range(1, 10):
    print("\n-= 第", i, "次 =-")
    longest_increasing_subseq()
```
