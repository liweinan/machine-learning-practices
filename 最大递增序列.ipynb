```python
import random


def longest_increasing_subseq():
    # 创建一个里面有10个随机数字的list
    random_list = []
    for i in range(10):
        random_list.append(random.randrange(1, 101, 1))
    print("原始列表: ", random_list)

    # marker用来标记上次一找到的sublist的位置
    marker = -1
    found = False

    target_list = []

    # 算法的思路如下：
    # 查看下一个元素，如果比之前的子序列里面的元素小，那么之前的子序列回退到比这个元素小的位置，然后在子序列里加入这个元素。
    for next_item in random_list:
        print("当前的递增序列：", target_list)
        print("下一个要添加的值：", next_item)
        # 使用enumerate函数，可以在遍历list的同时，得到每个元素的index值。
        for idx, _ in enumerate(target_list):
            print("当前的递增序列里要比对的值：", _)
            if _ > next_item and found is False:
                print("要添加的值更小！")
                print("元素位置：", idx)
                # 找到了之前的子序列里第一个大于下一个item value的item位置。
                marker = idx
                found = True
        if marker >= 0:
            target_list = target_list[:marker]
            marker = -1
            found = False
        target_list.append(next_item)

    print("最大递增序列：", target_list)


# 运行十次
for i in range(1, 10):
    print("\n-= 第", i, "次 =-")
    longest_increasing_subseq()
```
