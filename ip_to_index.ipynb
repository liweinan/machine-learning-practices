```python
def ip_to_num(ip):
    '''å°†ipè½¬åŒ–ä¸ºæ•´æ•°ï¼Œé¦–å…ˆåˆ†å‰²å‡ºIPå‰ä¸‰æ®µå¹¶è½¬åŒ–æˆintæ•°æ®ç±»åž‹ï¼Œä½¿ç”¨æ•´æ•°æ¢ç®—æ–¹æ³•è¿”å›žå¾—åˆ°IPå¯¹åº”çš„æ•´æ•°'''
    ips = ip.split(".")[0:3]
    # print(ip)
    # print(ips)
    return int(ips[2]) + int(ips[1]) * 256 + int(ips[0]) * 256 ** 2
```

```python
def find_index(interval, i):
    beg = 0
    end = len(interval) - 1
    cur = int((beg + end) / 2)

    while i < interval[cur][0] or i > interval[cur][1]:
        # print(cur, interval[cur])
        if i < interval[cur][0]:
            end = cur - 1
        elif i > interval[cur][1]:
            beg = cur + 1
        cur = int((beg + end) / 2)

    return cur
```
